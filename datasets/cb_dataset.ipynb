### EmotionLines

```python
import pandas as pd
import json
import csv

json_file_path = '..\EmotionLines\\friends_train.json'

with open(json_file_path, 'r', encoding='utf-8') as j:
     contents = json.loads(j.read())

# data = []
# for line in open('..\EmotionLines\\friends_train.json', 'r', newline=''):
#     data.append(json.loads(line))
```

```python
contents[0][0]["utterance"].encode('ascii', errors='ignore').decode("utf-8")
```

```python
file_name="friends_train.csv"
with open(file_name, "w",encoding='utf-8',newline='') as file:
    csv_file = csv.writer(file)
    header=['index', 'utterance', 'emotion']
    csv_file.writerow(header)
    for idx in range(len(contents)):
        for utt in range(len(contents[idx])):
            if idx%100==0:
                print(idx)
            csv_file.writerow([idx+1,contents[idx][utt]['utterance'], contents[idx][utt]['emotion']])
           
file.close()
```

```python
f = pd.read_csv('friends_train.csv')
f['emotion'].unique()
```
