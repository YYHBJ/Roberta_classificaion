```python
import json
import csv

data = []
for line in open('dataset_cn.json', 'r'):
    data.append(json.loads(line))
    
```

```python
with open("dataset_cn.csv", "w",encoding='utf-8',newline='') as file:
    csv_file = csv.writer(file)
    header=['index', 'ques_title', 'ques_content','ques_label', 'answer_content']
    csv_file.writerow(header)
    for idx in range(len(data)):
        if idx%10000==0:
            print(idx)
        csv_file.writerow([idx+1,data[idx]['ques_info']['title'], data[idx]['ques_info']['content'], data[idx]['ques_info']['ques_label'],''])
        for st in range(len(data[idx]['answers_info'])):
                csv_file.writerow([idx+1, "", "", data[idx]['ques_info']['ques_label'], data[idx]['answers_info'][st]['content']])
file.close()
```
